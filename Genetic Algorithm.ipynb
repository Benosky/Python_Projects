```python
import math
import scipy
import csv


#Import Tesla csv file to my computer and delete the title and every other unwanted columns
#leaving only the date and the close price columns.

tesla = open('tesla.csv','r')

tsl_pr= csv.reader(tesla)
date = []
prices = []

for line in tsl_pr:
    date.append(line[0])
    prices.append(float(line[1]))
    
prices

X=prices

sample_size=len(X) #Sample size
a = 1 #The constant in the fitness

def chroms(X):
    'returns the sets of prices that satify the given condition.'
    t=0
    chroms=[]
    for t in range(len(X)-36): #We make provision for chromo not to go out of bounds by 
        if X[t+7]<=254 and X[t+19]>=250 and X[t+23]<=248 and X[t+35]>=259: #using (len(X)-36)
            chromo = [X[t+7],X[t+14],X[t+19],X[t+23],X[t+26],X[t+33],X[t+35]] 
            chroms.append(chromo)
    return chroms


def genePop(X): #Population
    """Returns the distribution of all the prices that satify 
    the given condition: chromos"""
    t=0
    chrom_pop = [] 
    for t in range(len(X)-36):
        if X[t+7]<=254 and X[t+19]>=250 and X[t+23]<=248 and X[t+35]>=259:
            chromo = [X[t+7],X[t+19],X[t+23],X[t+35]] 
            chrom_pop = chrom_pop + chromo
            t+=1
    return chrom_pop


def eval_func(X): #Fitness
    """determining the fitness of each chromosome"""
    t=0
    a=1
    rank_chroms=[]
    for t in range(len(chroms(X))):
        rank=[-(math.log(scipy.std(chroms(X)[t])/scipy.std(genePop(X))))- (a/(len(genePop(X))))]
        rank_chroms=rank_chroms+rank
        t+=1
    return rank_chroms



        
#Create a dcitionary to tie the ranks to their respective chromosomes of prices
def dicRankChroms(X):
    dic_chrm={}
    for i in range (len(eval_func(X))):
        dic_chrm[eval_func(X)[i]]=chroms(X)[i]
        i+=1
    return dic_chrm

#Selecting the minimum rank/fitness 

def select_MinRnk(X):
    A=eval_func(X)
    min = A[0]
    for i in range(1,len(A)):
        if min > A[i]:
            min = A[i]
    return min

        
#Discarding the lowest-fitness chromosome and returning the rest as the new ranked chromosomes

def best_fit_chrms(X):
    R = eval_func(X)
    NewRankdChrms = []
    for i in range(len(R)):
        minR = select_MinRnk(X)
        del dicRankChroms(X)[minR]
        NewRankdChrms.append(dicRankChroms(X)[R[i]])
    return NewRankdChrms

#Applying crossover operatos to the selected parents(new ranked chromosomes) 
#and generating offsprings

def parent1(X): #Defining one set of parents (parent1) to aid pairing
    parent1, p = [],[]
    for i in range (0, len(best_fit_chrms(X)), 2):
        p = [best_fit_chrms(X)[i]]
        parent1 = parent1 + p
        i+=1
    return parent1


def parent2(X): #Defining the second set of parents (parent2) to aid pairing
    parent2, p = [],[]
    for i in range (1, len(best_fit_chrms(X)), 2):
        p = [best_fit_chrms(X)[i]]
        parent2= parent2 + p
        i+=1
    return parent2
    
        
#Generating the next generation of prices

def child1(X): #Defining the one set of children (child 1) from each pair of parents
    for t in range(len(parent1(X))):
        parent1(X)[t][2] = parent2(X)[t][2]
        child1 = parent1(X)
    return child1


def child2(X): #Defining the other set of children (child 2) from each pair of parents
    for t in range(len(parent2(X))):
        parent2(X)[t][:2] = parent1(X)[t][:2]
        parent2(X)[t][3:] = parent1(X)[t][3:]
        child2 = parent2(X)
    return child2


#Recombining the children to create a future generation (forecast) of prices.

def forecast(X):
    pop_child1, pop_child2 = [],[]
    for i in range(len(child1(X))):
        new_child1=child1(X)[i]+child1(X)[i+1]
        new_child2=child2(X)[i]+child2(X)[i+1]
        pop_child1=pop_child1 + new_child1
        pop_child2=pop_child2 + new_child2
        new_gen = pop_child1 + pop_child2
        i+=1
    return new_gen
        
print(forecast(X))        
        
    
                                                       
   

```
