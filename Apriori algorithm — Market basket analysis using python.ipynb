The Retailer of a retail store is trying to find out an association rule between 20 items, to figure out which items are more often bought together so that he can keep the items together in order to increase sales.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

#"/Users/benjamin/Desktop/red_car.jpeg"

dataset = pd.read_csv("/Users/benjamin/Market_Basket_Optimisation.csv", header = None)
dataset.head()
```

```python
#Convert Pandas DataFrame into a list of lists.

transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
#Build the Apriori model.

rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
#Print rules:
print(list(rules))
```
